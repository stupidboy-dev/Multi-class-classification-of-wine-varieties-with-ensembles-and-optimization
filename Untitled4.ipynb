```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.datasets import load_wine
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score, recall_score, accuracy_score,f1_score, precision_score, classification_report
from sklearn.model_selection import GridSearchCV, train_test_split
from imblearn.over_sampling import SMOTE

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data = load_wine()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target, name='target')
X = X.drop(['flavanoids', 'od280/od315_of_diluted_wines'], axis=1)
```

```python
scalar = StandardScaler()
scaled_x = scalar.fit_transform(X)
scaled_data = pd.DataFrame(scaled_x, columns=X.columns)
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_data, y, test_size=0.2,
                                                    random_state=42)
smote = SMOTE()
X_train, y_train = smote.fit_resample(X_train ,y_train)
```

```python
logst = LogisticRegression(random_state=42)
params = [
    {
        'penalty': ['l1', 'l2'],
        'C': [0.001, 0.01, 0.1, 1, 10, 100],
        'solver': ['liblinear', 'saga'],
        'class_weight': [None, 'balanced']
    },
    {
        'penalty': ['elasticnet'],
        'C': [0.001, 0.01, 0.1, 1, 10, 100],
        'solver': ['saga'],
        'l1_ratio': [0.3, 0.5, 0.7],
        'class_weight': [None, 'balanced']
    }
]
search = GridSearchCV(
        estimator=logst,
        param_grid=params,
        scoring='accuracy',
        cv=5,
        n_jobs=-1
)
search.fit(X_train, y_train)
best_log = search.best_estimator_
proba = search.predict_proba(X_test)
pred = best_log.predict(X_test)


print(classification_report(y_test, pred))
```

```python
rf = RandomForestClassifier(random_state=42)
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, None],
    'min_samples_split': [2, 5, 10],
    'max_features': ['sqrt', 'log2']
}
search_rf = GridSearchCV(estimator=rf,param_grid=param_grid,cv=5)
search_rf.fit(X_train, y_train)
best_rf = search_rf.best_estimator_
pred = best_rf.predict(X_test)
print(classification_report(y_test,pred))
```

```python
xgb = XGBClassifier(random_state=42)
param_grid = {
    'learning_rate': [0.05, 0.1],
    'n_estimators': [100, 200],
    'max_depth': [3, 5],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0],
    'reg_lambda': [1, 5],
    'gamma': [0, 0.1]
}

model = XGBClassifier()
search_x = GridSearchCV(model, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
search_x.fit(X_train, y_train)
best_xgb = search_x.best_estimator_
pred = best_xgb.predict(X_test)
print(classification_report(y_test,pred))
```
